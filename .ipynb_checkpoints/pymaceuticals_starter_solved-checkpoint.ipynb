### Analysis

Pymaceuticals, Inc. has recently began screening for potential treatments for squamous cell carcinoma (SCC), a commonly occurring form of skin cancer. This study was to compare the performance of Pymaceuticalsâ€™ drug of interest, Capomulin, against the other treatment regimens.

This study and analysis of the results aim to provide insights on the effectiveness of Capomulin and 9 other tested anti-cancer medications reducing tumors. This study was conducted on 249 mice, 51% male and 49% female. Over the course of 45 days, groups of mice were under different treament regimes, 25 mice per group. Every fifth day tumor development was observed and measured
Mice were distributed in groups of about 25 mice. Each group was treated with a different medication.

Analysis Takeaways

Capomulin is likely to be an effective medication to reduce a tumor's size. Mice treted with Capomilin and Rumican had the greater reduction in tumors's size compared to mice treated with other medications. All mice entered the study with a 45 mm3 tumors, and at the end of the of the study, the median volume of the tumors was 38.12 mm3 for mice treated with Capomulin and 40.65 for mice treted with Ramicane. Other treatment regimes had lower impact on reducing the size of tumors.

Prolonged use of Capomulin linked to stronger tumor reduction. A continues reduction on tumors' sizes was observed across all the mice medicate with Capomulin.

Mause weight is related to tumor size. Larger mice had bigger tumor sizes throughout the study. The correlation between mouse weight and the average tumor volume is 0.84. While there is no indication that patients that are overweight will develop bigger tumors, this finding suggest that weight control could be benefitial during skin cancer treatment with Capumulin or other treatment regimes. Further analysis will need to be conducted to confirm this hypothesis. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_data_combined = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])
# Display the data table for preview

mouse_data_combined.head()
```

```python
# Checking the number of mice.
number_mice = len(mouse_data_combined["Mouse ID"].unique())
print(number_mice)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouse_time = mouse_data_combined[["Mouse ID","Timepoint"]].copy()
dup_loc= mouse_time.duplicated()
mouse_data_combined["Duplicate?"] = dup_loc
duplicated_mouse = mouse_data_combined.loc[mouse_data_combined["Duplicate?"],["Mouse ID"]]
duplicated_mouse["Mouse ID"].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicated_mouse_data = mouse_data_combined.loc[mouse_data_combined["Mouse ID"].isin(duplicated_mouse["Mouse ID"]), :]
duplicated_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_mouse_data_combined = mouse_data_combined[~mouse_data_combined["Mouse ID"].isin(duplicated_mouse["Mouse ID"])]
clean_mouse_data_combined.head()
```

```python
# Checking the number of mice in the clean DataFrame.
print(len(clean_mouse_data_combined["Mouse ID"].unique()))
```

```python
 clean_mouse_data_combined.groupby(["Drug Regimen"])["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume.
#mean
per_drug_regimen_mean_tumor_volume = clean_mouse_data_combined.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
#median
per_drug_regimen_median_tumor_volume = clean_mouse_data_combined.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
#variance
per_drug_regimen_var_tumor_volume = clean_mouse_data_combined.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
#Standard Dev
per_drug_regimen_std_tumor_volume = clean_mouse_data_combined.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
#Sem
per_drug_regimen_sem_tumor_volume = clean_mouse_data_combined.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary DataFrame.
drug_regimen_summary = pd.DataFrame({
                                "Mean Tumor Volume": per_drug_regimen_mean_tumor_volume,
                                "Median Tumor Volume": per_drug_regimen_median_tumor_volume,
                                "Tumor Volume Variance": per_drug_regimen_var_tumor_volume,
                                "Tumor Volume Std. Dev.": per_drug_regimen_std_tumor_volume,  
                                "Tumor Volume Std. Err.": per_drug_regimen_sem_tumor_volume
                                })

# Display the DataFrame
drug_regimen_summary
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
drug_regimen_grouped = clean_mouse_data_combined.groupby(["Drug Regimen"]).agg({"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]})
drug_regimen_grouped
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

max_timepoint_mouse = clean_mouse_data_combined.groupby(["Mouse ID"]).max()["Timepoint"]

last_timepoint_mouse = max_timepoint_mouse.to_frame()
last_timepoint_mouse.reset_index()


last_timepoint_mouse_all = last_timepoint_mouse.merge(clean_mouse_data_combined, how='inner', left_on=['Mouse ID','Timepoint'], right_on=['Mouse ID', 'Timepoint'])

last_timepoint_mouse_all

```

```python
# Put treatments into a list for for loop (and later for plot labels)

treatment_groups= ["Capomulin", "Ramicane", "Infubinol","Ceftamin"]


# Create empty list to fill with tumor vol data (for plotting)
vol_dict = {}
treatment = []
vol_data = []

for drug in range(len(treatment_groups)):
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    treatment = treatment_groups[drug]
    vol_data = last_timepoint_mouse_all.loc[(last_timepoint_mouse_all["Drug Regimen"] == treatment_groups[drug] ), 
                                         ["Tumor Volume (mm3)"]]

    iqr_data= vol_data["Tumor Volume (mm3)"]
    # Calculate the IQR and quantitatively determine if there are any potential outliers. 
    quartiles = iqr_data.quantile([0.25,0.5,0.75])
    # Determine outliers using upper and lower bounds
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    print(f"Treatment: {treatment}")
    print(f"The lower quartile of temperatures is: {lowerq}")
    print(f"The upper quartile of tumor volume (mm3) is: {upperq}")
    print(f"The interquartile range of tumor volume is: {iqr}")
    print(f"The the median of tumor volume is: {quartiles[0.5]} ")
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
    
     # Directory storing treatment and tumor volume valaues
    vol_dict[treatment] = vol_data["Tumor Volume (mm3)"].values.tolist()
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
box_plot_data = pd.DataFrame.from_dict(vol_dict)

# Plot the dataframe

ax = box_plot_data[treatment_groups].plot(kind='box', title='boxplot')
ax.set_ylabel("Final Tumor Volume (mm3)")
#treatment_groups
# Display the plot
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

line_data = clean_mouse_data_combined.loc[(clean_mouse_data_combined["Drug Regimen"] == "Capomulin") & ( clean_mouse_data_combined["Mouse ID"] == "l509"  ), 
                                         ["Timepoint","Tumor Volume (mm3)"]]
#line_data
lines = line_data.plot.line(x='Timepoint', y='Tumor Volume (mm3)')
lines.set_xlabel("Timepoint (days)")
lines.set_ylabel("Tumor Volume (mm3)")
plt.title("Capomulin treatment of mouse l509")
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
scatter_data = clean_mouse_data_combined.loc[(clean_mouse_data_combined["Drug Regimen"] == "Capomulin"), 
                                         ["Tumor Volume (mm3)", "Weight (g)", "Mouse ID"]]

scatter_data_avgvolume = scatter_data.groupby(["Mouse ID"]).mean()["Tumor Volume (mm3)"]

scatter_data_avgvolume_df = scatter_data_avgvolume.to_frame()

scatter_data_weight = scatter_data[['Mouse ID', 'Weight (g)']].copy()

scatter_data_weight_ndp = scatter_data_weight.drop_duplicates()

scatter_plot_data= pd.merge(scatter_data_weight_ndp, scatter_data_avgvolume_df, how="left", on=["Mouse ID", "Mouse ID"])

scatter_plot = scatter_plot_data.plot.scatter(x='Weight (g)', y='Tumor Volume (mm3)')

scatter_plot.set_ylabel("Average Tumor Volume (mm3)")

plt.title("Average Tumor Volume (mm3) vs Mouse Weight")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
x = np.array(scatter_plot_data["Weight (g)"].values.tolist())
y = np.array(scatter_plot_data["Tumor Volume (mm3)"].values.tolist())

corr, p = st.stats.pearsonr(x, y)
print(f"The correlation between mouse weight and the average tumor volume is {corr}.")

res = st.stats.linregress(x, y)
plt.plot(x, y, 'o')

                    
plt.plot(x, (res.intercept + res.slope*x), 'r')
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

plt.show()
```
